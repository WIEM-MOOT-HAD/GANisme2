## map cities

```python
#regex to get cities names: 
import re 
city_pattern=', (.*)'
text = df_painting["LOCATION"].to_list()
cities = []
for location in text:
  n_city = re.findall(city_pattern, location, flags=re.IGNORECASE)
  cities.append(n_city)
```

```python
flatten_list=[]
for sublist in cities:
    flatten_list.extend(sublist)


list_of_cities = ["Cambridge" if item =="Harvard University, Cambridge" else item for item in flatten_list]

list_of_cities = ["winterthur" if item =="Winthertur"else item for item in list_of_cities]
list_of_cities = ["Rome" if item =="Vatican Palace, Rome" else item for item in list_of_cities]
list_of_cities = ["Vatican" if item =="Palazzi Pontifici, Vatican" else item for item in list_of_cities]
list_of_cities = ["Madrid" if item =="Vienna, Academia de San Fernando, Madrid, Private collection" else item for item in list_of_cities]
list_of_cities = ["Denver" if item =="Vienna, Art Museum, Denver, Private collection" else item for item in list_of_cities]
list_of_cities = ["Bologna" if item =="San Giacomo Maggiore, Bologna" else item for item in list_of_cities]
list_of_cities = ["Cittadella" if item =="Parish Church, Cittadella" else item for item in list_of_cities]
list_of_cities = ["Rome" if item =="Vatican City, Rome" else item for item in list_of_cities]
list_of_cities = ["Florence" if item =="Galleria Palatina (Palazzo Pitti), Florence" else item for item in list_of_cities]
list_of_cities = ["Rome" if item =="San Luigi dei Francesi, Rome" else item for item in list_of_cities]
list_of_cities = ["Rome" if item =="Villa Boncompagni Ludovisi, Rome" else item for item in list_of_cities]
list_of_cities = ["Mantua" if item =="San Maurizio, Mantua" else item for item in list_of_cities]
list_of_cities = ["Linz" if item =="Wolfgang-Gurlitt-Museum, Linz" else item for item in list_of_cities]
list_of_cities = ["Siena" if item =="Spedale di Santa Maria della Scala, Siena" else item for item in list_of_cities]

```

```python
# Import the required library
from geopy.geocoders import Nominatim

# Initialize Nominatim API
geolocator = Nominatim(user_agent="MyApp")
latitudes=[]
longitudes=[]
for city in list_of_cities: 
  print(city)
  location = geolocator.geocode(city)
  latitudes.append(location.latitude)
  print("len latitudes=", len(latitudes))
  longitudes.append(location.longitude)
  print("len latitudes=", len(longitudes))

  #print("The latitude of the location is: ", location.latitude)
  #print("The longitude of the location is: ", location.longitude)
```

```python
#setting center for our map
center = [20.593684,78.96288]

# creating map
map = folium.Map(location = center, zoom_start = 2)
```
